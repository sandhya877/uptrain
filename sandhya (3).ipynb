```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from time import sleep

df = pd.read_csv('Input.csv')

for index, row in df.iterrows():
    id = row['URL_ID']
    url = row['URL']
    print(id)
    print(url)
    trails = 1
    while trails <= 5:
        try:
            r = requests.get(url)
            print(r.status_code)
            if r.status_code == 200:
                soup = BeautifulSoup(r.content, "html.parser")
                title = soup.title.string
                mc = soup.find("div", class_="td-post-content")
                body = "\n".join([p.get_text() for p in mc.find_all(["p", "li"])])
                with open(f'{id}_title.txt', 'w') as f:
                    f.write(str(title))
                with open(f'{id}_body.txt', 'w') as f:
                    f.write(str(body))
                break
            elif r.status_code == 404:
                print(f'{id} Page doesnt exist: {url}')
                continue
        except:
            trails += 1
            sleep(5)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
nltk.download('punkt')
nltk.download('cmudict')
```

```python
import re

def keep_alphabets(text):
    pattern = re.compile('[^a-zA-Z]')
    clean_text = re.sub(pattern, ' ', text)
    
    return clean_text
```

```python
def clean_text(text):
    if not isinstance(text, (str, bytes)):
        try:
            text = str(text)
        except:
            text = str(text.encode('utf-8'))
    
    return keep_alphabets(text)
```

```python
def text_analysis(row):
    id = row['URL_ID']
    print('id: ', id)
    t_fname = str(id)+'_title.txt'
    b_fname = str(id)+'_body.txt'
    try:
        with open('data/'+t_fname) as f:
            head = [line.strip() for line in f.readlines()]

        with open('data/'+b_fname) as f:
            body = [line.strip() for line in f.readlines()]

        text = clean_text(str(head) + str(body))
    except:
        text = ' '
    print('text: ', text[0:100])
    # 1. Positive Score
    positive_words = ['improve', 'good', 'great', 'excellent', 'positive']
    positive_score = 0

    # 2. Negative Score
    negative_words = ['bad', 'negative', 'terrible', 'awful']
    negative_score = 0

    # Tokenize the words in the text
    words = word_tokenize(text)

    # 3. Polarity Score
    for word in words:
        if word in positive_words:
            positive_score += 1
        elif word in negative_words:
            negative_score -= 1
            
    row['POSITIVE SCORE'] = positive_score
    row['NEGATIVE SCORE'] = negative_score
    polarity_score = 0
    if positive_score > 0 and negative_score > 0:
        polarity_score = (positive_score - negative_score) / (positive_score + negative_score)
    row['POLARITY SCORE'] = polarity_score
    print('+ve, -ve, polarity: ', positive_score, negative_score, polarity_score)
    # 4. Subjectivity Score
    total_words = len(words)
    subjectivity_score = 0
    if total_words > 0:
        subjectivity_score = (positive_score + negative_score) / (total_words)
    row['SUBJECTIVITY SCORE'] = subjectivity_score
    print('subjectivity_score: ', subjectivity_score)
    # 5. Average Sentence Length
    sentences = sent_tokenize(text)
    total_sentences = len(sentences)
    avg_sentence_length = 0
    if total_sentences > 0:
        avg_sentence_length = total_words / total_sentences
    row['AVG SENTENCE LENGTH'] = avg_sentence_length
    print('avg_sentence_length: ', avg_sentence_length)
    # 6. Percentage of Complex Words
    complex_words = 0
#     for word in words:
#         try:
#             if len([syllable for syllable in nltk.corpus.cmudict.dict()[word.lower()]]) >= 3:
#                 complex_words += 1
#         except:
#             pass
    
    percentage_of_complex_words = 0
    if total_words > 0:
        percentage_of_complex_words = complex_words / total_words
    row['PERCENTAGE OF COMPLEX WORDS'] = percentage_of_complex_words
    print('percentage_of_complex_words:', percentage_of_complex_words)
    # 7. Fog Index
    fog_index = 0.4 * (avg_sentence_length + percentage_of_complex_words)
    row['FOG INDEX'] = fog_index
    print('fog_index:', fog_index)
    # 8. Average Number of Words Per Sentence
    avg_number_of_words_per_sentence = 0
    if total_sentences:
        avg_number_of_words_per_sentence = total_words / total_sentences
    row['AVG NUMBER OF WORDS PER SENTENCE'] = avg_number_of_words_per_sentence
    print('avg_number_of_words_per_sentence:', avg_number_of_words_per_sentence)
    # 9. Complex Word Count
    row['COMPLEX WORD COUNT'] = complex_words
    print('complex_words:', complex_words)
    # 10. Word Count
    row['WORD COUNT'] = total_words
    print('total_words:', total_words)
    # 11. Syllable Per Word    
    syllable_per_word = []
#     for word in words:
#         try:
#             syllable_count = len([syllable for syllable in nltk.corpus.cmudict.dict()[word.lower()]])
#             syllable_per_word.append(syllable_count)
#         except KeyError:
#             pass

    row['SYLLABLE PER WORD'] = syllable_per_word
    print('syllable_per_word:', syllable_per_word)
    
    # 12. Personal Pronouns
    personal_pronouns = ['I', 'we', 'my', 'ours', 'us']
    personal_pronoun_count = 0
    for word in words:
        if word in personal_pronouns:
            personal_pronoun_count += 1
    row['PERSONAL PRONOUNS'] = personal_pronoun_count
    print('personal_pronoun_count:', personal_pronoun_count)

    # 13. Average Word Length
    avg_word_length = 0
    if total_words:
        avg_word_length = sum([len(word) for word in words]) / total_words
    row['AVG WORD LENGTH'] = avg_word_length
    print('avg_word_length:', avg_word_length)
    return row
```

```python
df = df.apply(text_analysis, axis=1)
```
